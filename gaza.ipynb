```python
from GoogleNews import GoogleNews
import pandas as pd
from datetime import datetime, timedelta

# Create a list to store results
all_results = []

# Create a list of keywords or topics you are interested in
keywords_list = ["palestine", "conflict", "middle east", "peace process", "humanitarian"]

# Define the date range (past 3 months)
end_date = datetime.now()
start_date = end_date - timedelta(days=90)

# Loop through date ranges and make queries
current_date = start_date
while current_date <= end_date:
    gnews = GoogleNews(start=current_date.strftime('%m/%d/%Y'), end=(current_date + timedelta(days=7)).strftime('%m/%d/%Y'))
    for keyword in keywords_list:
        gnews.search(keyword)
        result = gnews.result()
        all_results.extend(result)
    current_date += timedelta(days=7)

# Convert the results to a DataFrame
data = pd.DataFrame.from_dict(all_results)
data.head(20)  # Display the first 20 rows as an example
```

```python
from GoogleNews import GoogleNews
import pandas as pd
from datetime import datetime, timedelta
import time
# Create a list to store results
all_results = []

# Create a list of keywords or topics you are interested in
keywords_list = ["palestine", "conflict", "middle east", "Hamas", "Gaza","israel"]

# Define the date range (from January 1, 2023, to December 31, 2023)
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 2, 1)

# Loop through date ranges and make queries for each day
current_date = start_date
while current_date <= end_date:
    gnews = GoogleNews(start=current_date.strftime('%m/%d/%Y'), end=current_date.strftime('%m/%d/%Y'))
    for keyword in keywords_list:
        gnews.search(keyword)
        result = gnews.result()
        all_results.extend(result)

    current_date += timedelta(days=1)

    # Add a delay between requests to avoid rate limiting
    time.sleep(1)  # You can adjust the delay as needed
# Convert the results to a DataFrame
data = pd.DataFrame.from_dict(all_results)

```
